```python
# fuc to drop category data

def drop_cat_outliers(data, column, count_threshold):
    
    
    series = data[column].value_counts()
    filt = series < count_threshold
    outliers = filt[filt].index
    
    func = lambda x: x in outliers
    filt = data[column].apply(func)
    outlier_indices = filt[filt].index
    
    data.drop(outlier_indices, inplace = True)
    data.reset_index(drop = True, inplace = True)
    
    is_unchanged = data.shape[0] == data.shape[0]
    
    return is_unchanged, data
```

```python
# drop cont data 
def drop_cont_outliers(data, column, business_logic_whiskers = (0, None)):
    
    data_ = data.copy(deep = True)
    
    q1 = data_[column].quantile(0.25)
    q3 = data_[column].quantile(0.75)
    iqr = q3 - q1
    
    business_lower, business_upper = business_logic_whiskers
    lower_whisker = q1 - 1.5 * iqr
    if business_lower != None and business_lower > lower_whisker:
        lower_whisker = business_lower
            
    upper_whisker = q3 + 1.5 * iqr
    if business_upper != None and business_upper < upper_whisker:
        upper_whisker = business_upper
        
    lower_filt = data_[column] < lower_whisker
    upper_filt = data_[column] > upper_whisker
    filt = lower_filt | upper_filt
    outlier_indices = filt[filt].index
    
    data.drop(outlier_indices, inplace = True)
    data.reset_index(drop = True, inplace = True)
    
    is_unchanged = data.shape[0] == data.shape[0]
    
    return is_unchanged, data
```
